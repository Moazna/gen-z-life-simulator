```python
def lifeSimulator(user):
    stats = {"energy": 100, "points": 0, "mood": "Happy"}
    moods_playlist = {
        "Happy": ["K-pop", "Lo-fi Chill", "Pop Hits"],
        "Sad": ["Acoustic Sad Songs", "Lo-fi Slow Vibes"],
        "Angry": ["Rock", "Heavy Metal"],
        "Tired": ["Lo-fi Sleepy Beats", "Soft Instrumentals"],
        "Excited": ["Upbeat Pop", "EDM"]
    }

    def showStatus():
        print(f"\nğŸ’– {user['name']}'s Stats ğŸ’–")
        print(f"Energy: {stats['energy']}")
        print(f"Points: {stats['points']}")
        print(f"Mood: {stats['mood']}")
        print(f"ğŸµ Mood Playlist: {', '.join(moods_playlist.get(stats['mood'], ['Random Vibes']))}\n")

    def updateStats(activity):
        if activity == "study":
            stats["energy"] -= 20
            stats["points"] += 10
            stats["mood"] = "Tired"
        elif activity == "eat":
            stats["energy"] += 15
            stats["mood"] = "Happy"
        elif activity == "play":
            stats["energy"] -= 10
            stats["points"] += 5
            stats["mood"] = "Excited"
        elif activity == "sleep":
            stats["energy"] = 100
            stats["mood"] = "Happy"
        else:
            print("ğŸ¤” Unknown activity...")
        stats["energy"] = max(0, min(100, stats["energy"]))  # cap energy

    print("\nğŸ® Welcome to Life Simulator Mode! ğŸ®")
    while True:
        showStatus()
        print("Choose an activity: study / eat / play / sleep / quit")
        activity = input("Your choice: ").lower().strip()
        if activity == "quit":
            print(f"\nğŸ’– Game Over! You scored {stats['points']} points with {stats['energy']} energy left. Mood: {stats['mood']} ğŸ’–")
            break
        updateStats(activity)
```
